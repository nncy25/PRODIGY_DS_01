```python
import pandas as pd
import random

# Your 100 names
names = [
    "Olivia", "Liam", "Emma", "Noah", "Ava", "Elijah", "Sophia", "James", "Isabella", "William",
    "Mia", "Benjamin", "Charlotte", "Lucas", "Amelia", "Henry", "Harper", "Alexander", "Evelyn", "Michael",
    "Abigail", "Daniel", "Ella", "Matthew", "Lily", "Jackson", "Scarlett", "Sebastian", "Grace", "Jack",
    "Chloe", "Aiden", "Zoey", "Owen", "Penelope", "Samuel", "Layla", "Carter", "Riley", "Wyatt",
    "Aria", "John", "Nora", "David", "Camila", "Joseph", "Aurora", "Leo", "Ellie", "Julian",
    "Mila", "Luke", "Paisley", "Grayson", "Skylar", "Isaac", "Brooklyn", "Jayden", "Bella", "Gabriel",
    "Claire", "Anthony", "Savannah", "Dylan", "Addison", "Lincoln", "Aubrey", "Ezra", "Ellie", "Thomas",
    "Stella", "Charles", "Natalie", "Christopher", "Hazel", "Jaxon", "Violet", "Maverick", "Lucy", "Josiah",
    "Hannah", "Hudson", "Lillian", "Nathan", "Leah", "Caleb", "Zoe", "Elias", "Audrey", "Aaron",
    "Allison", "Ryan", "Samantha", "Adrian", "Sadie", "Nolan", "Eva", "Christian", "Kennedy", "Jonathan"
]

# Random ages between 21 and 45
ages = [random.randint(21, 45) for _ in range(100)]

# Random genders
genders = [random.choice(['Male', 'Female']) for _ in range(100)]

# Create DataFrame
df = pd.DataFrame({
    'Name': names,
    'Age': ages,
    'Gender': genders
})

# Save as CSV file
df.to_csv("people_dataset.csv", index=False)

# Print the whole dataset (all 100 rows)

pd.set_option('display.max_rows', None)
print(df)

#AGE GROUP DISTRIBUTION CHART

import matplotlib.pyplot as plt

# Create age bins and labels
bins = [20, 25, 30, 35, 40, 45]
labels = ['21-25', '26-30', '31-35', '36-40', '41-45']

# Create a new column 'Age Group' based on bins
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Count how many people are in each age group
age_group_counts = df['Age Group'].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(8,5))
age_group_counts.plot(kind='bar', color='lightgreen')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of People')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# GENDER DISTRIBUTION CHART

# Count number of people by gender
gender_counts = df['Gender'].value_counts()

# Plot bar chart
plt.figure(figsize=(6,4))
gender_counts.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of People')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```
